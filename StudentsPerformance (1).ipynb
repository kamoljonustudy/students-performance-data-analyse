```python
df['Total Score'] = (df['math score'] + df['reading score'] + df['writing score'])/3
```

```python
df.head()
```

```python
df.groupby(by='race/ethnicity')['Total Score'].mean()
```

```python
df['race/ethnicity'].value_counts()
```

```python
df_high = df[df['Total Score'] >= 80]
```

```python
df_high['race/ethnicity'].count()
```

```python
df_high.groupby(by='race/ethnicity')[['math score', 'reading score', 'writing score']].count()
```
